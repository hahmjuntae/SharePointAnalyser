```python
# 필요한 라이브러리 임포트
import pandas as pd
import json
from datetime import datetime, timedelta
from google.colab import files

# KST 시간 포맷으로 변환하는 함수
def convert_utc_to_kst(utc_time_str):
    utc_time = datetime.strptime(utc_time_str, '%Y-%m-%dT%H:%M:%S')
    kst_time = utc_time + timedelta(hours=9)

    # 한국어 요일 변환
    day_korean = {
        "Monday": "월", "Tuesday": "화", "Wednesday": "수",
        "Thursday": "목", "Friday": "금", "Saturday": "토", "Sunday": "일"
    }
    day_of_week = day_korean[kst_time.strftime('%A')]
    return kst_time.strftime(f'%Y년 %m월 %d일({day_of_week}) %H:%M:%S')

# CSV 파일을 Colab에 업로드하기
print("CSV 파일을 업로드하세요.")
uploaded = files.upload()

# 업로드한 파일을 DataFrame으로 읽기
file_name = list(uploaded.keys())[0]  # 업로드한 파일명 가져오기
df = pd.read_csv(file_name)

# AuditData 컬럼에서 JSON 데이터를 추출하는 함수
def extract_audit_data(df):
    audit_data_list = []
    for _, row in df.iterrows():
        audit_data = json.loads(row['AuditData'])
        audit_data_list.append(audit_data)
    audit_df = pd.DataFrame(audit_data_list)
    return audit_df

# 필터링 및 포맷 적용 함수
def filter_and_format_data(df):
    keys_to_extract = [
        "Operation", "Workload", "CreationTime", "UserId", "ClientIP", "UserAgent", "Platform", "IsManagedDevice",
        "SiteUrl", "SourceRelativeUrl", "SourceFileName", "SourceFileExtension"
    ]
    filtered_df = df[keys_to_extract]
    filtered_df['CreationTime'] = filtered_df['CreationTime'].apply(convert_utc_to_kst)
    return filtered_df

# JSON 필드 추출 및 필터링 데이터 생성
df = extract_audit_data(df)
filtered_df = filter_and_format_data(df)

# 필터링된 데이터 확인
print("### 필터링된 데이터 미리보기 ###")
filtered_df.head()

# 필터링된 데이터를 Excel 파일로 저장
filtered_df.to_excel("filtered_audit_data.xlsx", index=False)
print("필터링된 데이터를 'filtered_audit_data.xlsx' 파일로 저장했습니다.")

# 저장된 파일 다운로드
files.download("filtered_audit_data.xlsx")
```
